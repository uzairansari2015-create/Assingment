1. Write a Python program to create a class representing a Circle. Include methods to calculate its area and perimeter.

```python
# Create a class
import math
class Circle():
    def __init__(self, radius): # Constructor
        self.radius=radius
    def calculate_area(self): #method
        return math.pi * self.radius**2
    def calculate_perimeter(self):
        return 2 * math.pi * self.radius
        pass
```

```python
#user input
radius = float(input("Enter the radius"))
```

```python
#create object
Circle = Circle(radius)
```

```python
print("\nCircle details")
print("----------------")
print("Radius", radius)
print("Area", round(Circle.calculate_area(),2))
print("Perimeter", round(Circle.calculate_perimeter(),2))
```

2. Write a Python program to create a person class. Include attributes like name, country and date of birth. Implement a method to determine the person's age.


```python
from datetime import date

class Person():
    def __init__(self, name,country,birth_year,birth_month,birth_day):
        self.name = name
        self.country = country
        self.birth_date = date(birth_year, birth_month, birth_day)
    def calculate_age(self):
        today = date.today()
        age = today.year - self.birth_date.year

        # check if birthday occurred this year
        if(today.month, today.day) < (self.birth_date.month, self.birth_date.day):
            age-=1
        return age
```

```python
person = Person("Uzair", "Pakistan", 1993, 5,29)
```

```python
print("Name:", person.name)
print("country:", person.country)
print("Age:", person.calculate_age(), "year")
```

3. Write a Python program to create a calculator class. Include methods for basic arithmetic operations.

```python
class Calculator():
    #For Addition
    def add(self, num1, num2):
        return num1 + num2

    #For Subtraction
    def subtract(self, num1, num2):
        return num1 - num2

    #For Multiplication
    def multiply(self, num1, num2):
        return num1 * num2

    #For Division
    def divide(self, num1, num2):
        if num2 == 0:
            return "Division by zero is not allowed"
        return num1 / num2
```

```python
calc = Calculator()

num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))
```

```python
print("\nCalculator result")
print("-------------------")
print("Addition:", calc.add(num1, num2))
print("Subtraction:", calc.subtract(num1, num2))
print("Multiplication:", calc.multiply(num1, num2))
print("Division:", calc.divide(num1, num2))
```

6. Create a list of five numbers and find the largest and smallest number without using the max() & min() function.

```python
numbers = [25,15,50,5,10]

largest = numbers[0]
smallest = numbers[0]

for number in numbers:
    if number > largest:
        largest = number

    if number < smallest:
        smallest = number

print("List of Numbers", numbers)
print("Largest Numbers", largest)
print("Smalles Numbers", smallest)
    
```
