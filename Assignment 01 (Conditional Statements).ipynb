1) A company decided to give bonus of 5% to employee if his/her year of service is more than 5 years.
Ask user for their salary and year of service and print the net bonus amount.

```python
salary = float(input("Enter your salary"))
Year_of_service = int(input("Enter year of service"))

if Year_of_service>=5:
    bonus = salary * 0.05
    print(f"Your bonus amount is:", bonus)

else:
    bonus = 0
    print(f"You are not eligible for bonus")
print("Net bonus amount is:", bonus)
```

8. A school has following rules for grading system:

a. Below 25 - F

b. 25 to 45 - E

c. 45 to 50 - D

d. 50 to 60 - C

e. 60 to 80 - B

f. Above 80 - A

Ask user to enter marks and print the corresponding grade.

```python
marks = int(input("Enter your marks"))
if marks>=0 and marks<=25:
    print(f"your marks is {marks} and your grade is F")
elif marks>=25 and marks<=45:
    print(f"your marks is {marks} and your grade is E")
elif marks>=45 and marks<=50:
    print(f"your marks is {marks} and your grade is D")
elif marks>=50 and marks<=60:
    print(f"your marks is {marks} and your grade is C")
elif marks>=60 and marks<=80:
    print(f"your marks is {marks} and your grade is B")
elif marks>=80:
    print(f"your marks is {marks} and your grade is A")
else:
    print(f"You entered an Invalid number")
```
