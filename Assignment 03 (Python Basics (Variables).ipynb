3. Write a program in which use all the operators we can use in Python

```python
a = 10
b = 5

print("Airthematic Operator")
print("---------------------")
print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Floor Division:", a // b)
print("Modulus:", a % b)
print("Exponent:", a ** b)

print("\nCompatison Operator")
print("---------------------")
print("a==b", a==b)
print("a!=b", a!=b)
print("a>b", a>b)
print("a<b", a<b)
print("a>=b", a>=b)
print("a<=b", a<=b)

print("\nAssignment Operator")
print("---------------------")
x = 15
print("Initial Value:", x)

x+=25
print("After +=:", x)

x-=3
print("After -=:", x)

x*=3
print("After *=:", x)

x/=3
print("After /=:", x)

x//=3
print("After //=:", x)

x%=3
print("After %=:", x)

x**=3
print("After **=:", x)

print("\nLogical Operators")
print("---------------------")
print((a>b) and (b>0))
print((a<b) and (b<0))

print("\nBitwise Operators")
print("---------------------")
print("a & b =", a & b)
print("a | b =", a | b)
print("a ^ b =", a ^ b)
print("~a =", ~a)
print("a << 1 =", a << 1)
print("a >> 1 =", a >> 1)
```
